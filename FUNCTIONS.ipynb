# 1.What is the difference between a function and a method in Python?
   - In Python, functions and methods are both blocks of reusable code, but they differ in how they are defined and used.

    ðŸ”¹ Function:
* A function is defined using the def keyword.  
* It is not associated with any object.
* You call a function directly using its name.
   
     ðŸ”¹ Method:
     *   A method is a function that is associated with an object.
     *  It is defined inside a class and can only be called through an instance (or the class, for class methods).
    *  It automatically takes the object (self) as the first parameter.
    









```python
# EXAMPLE

# Function example
def add(a, b):
    return a + b

result = add(3, 5)
print("Function result:", result)


# Method example
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
method_result = calc.add(3, 5)
print("Method result:", method_result)
```

```python
# 3. What are the different ways to define and call a function in Python?

   # 1. Standard Function (Positional Arguments)
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet("Alice", 25)

# 2. Default Arguments
def greet(name, age=18):
    print(f"Hello {name}, you are {age} years old.")

greet("Bob")           # Uses default value
greet("Charlie", 30)   # Overrides default

# 3. Keyword Arguments
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet(age=40, name="David")  # Order can be changed

# 4. Variable-Length Arguments (*args)
def add_numbers(*args):
    print("Sum:", sum(args))

add_numbers(1, 2, 3)
add_numbers(10, 20)

# 5. Keyword Variable-Length Arguments (**kwargs)
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Eve", age=28, city="Paris")

# 6. Lambda Function
square = lambda x: x * x
print("Square:", square(5))
```

# 5. What are iterators in Python and how do they differ from iterables?
  
  - Iterables are objects that can return an iterator, allowing you to loop over their elements (e.g., lists, tuples, strings).

   Iterators are objects that represent a stream of data; they keep track of the current position and produce the next item when you call next().

- Difference:

   - Iterable: Has an __iter__() method that returns an iterator.

   - Iterator: Has a __next__() method to get the next item and raises StopIteration when done.

```python
# EXAMPLE

my_list = [1, 2, 3]      # Iterable
my_iter = iter(my_list)  # Iterator

print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3
```

# 6. Explain the concept of generators in Python and how they are defined.

   - Generators in Python are special functions that produce a sequence of values lazily, one at a time, instead of returning them all at once. This makes them memory-efficient for large data.

   - How they work:

     - Use the yield keyword instead of return.

     - Each yield pauses the function, saving its state.

     - When resumed, it continues from where it left off.



```python
# EXAMPLE
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for number in count_up_to(3):
    print(number)
```

# 7. What are the advantages of using generators over regular functions?

  - Here are the main advantages of using generators over regular functions:

    1. Memory Efficiency
Generators produce items one at a time, so they use much less memory than functions that return full lists.

    2. Lazy Evaluation
Values are generated only when needed, which can improve performance when dealing with large data or infinite sequences.

   3. Represent Infinite Sequences
Generators can model infinite streams of data without running out of memory.

   4. Simpler Code for Iteration
Generators make it easy to write iterators without creating a class with __iter__() and __next__() methods.

    5. Pipeline Processing
Generators can be chained, enabling efficient processing of data streams in stages.

```python
# EXAMPLE
def fibonacci(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

for num in fibonacci(5):
    print(num)
```

# 9. Explain the purpose and usage of the `map()` function in Python.
  - The map() function applies a given function to each item of one or more iterables (like lists) and returns an iterator with the results.

  - Purpose:

    - To transform all items in an iterable by applying a function.

    - Useful for processing data without explicit loops.



```python
# EXAMPLE
numbers = [1, 2, 3, 4]
squares = map(lambda x: x ** 2, numbers)
print(list(squares))
```

# 10. What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?
  - map() applies a function to each item of an iterable and returns an iterator with the results.

- filter() applies a function that returns True or False to each item and returns only those items where the function returns True.

- reduce() applies a function cumulatively to the items of an iterable, reducing it to a single value.



```python
# EXAMPLE
from functools import reduce

# map: square each number
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))

# filter: keep only even numbers
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))

# reduce: sum all numbers
total = reduce(lambda x, y: x + y, numbers)
print(total)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.n1, self.n2 = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        else:
            val = self.n2
            self.n1, self.n2 = self.n2, self.n1 + self.n2
            self.count += 1
            return val

# Testing
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Given string
text = "Hello, how are you?"

# Function to check if a character is not a vowel
def not_vowel(char):
    return char.lower() not in 'aeiou'

# Use filter to remove vowels
filtered_chars = filter(not_vowel, text)

# Join the filtered characters back into a string
result = ''.join(filtered_chars)

print(result)
```
